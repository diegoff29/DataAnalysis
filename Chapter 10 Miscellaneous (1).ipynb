```python
#0=tails
#1=heads
from random import *
summax=0
for d in range(10000):
    coins=[randint(0,1) for b in range(200)]
    cons=[]
    a=0
    b=0
    while (a<len(coins)):
        if 0<=(a-1):
            if coins[a]==coins[a-1]:
                cons[b]+=1
            else:
                cons.append(1)
                b+=1
        else:
            cons.append(1)
        a+=1
    print(max(cons))
    summax=max(cons)+summax
print(summax/10000)
```
